```python
data = []
counterYolo =0
counterOF = 0
counterIteration = 0
while (counterYolo<dataYolo.shape[0] and counterOF<dataOF.shape[0]):
    if (dataYolo[counterYolo,0]<dataOF[counterOF,0]):
        # if data is -1 -> repace previous data
        for i  in range(dataYolo.shape[1]):
            if dataYolo[counterYolo,i] == -1:
                if counterIteration != 0:
                    dataYolo[counterYolo,i] = data[counterIteration-1][i]
        data.append(dataYolo[counterYolo,:].tolist())
        counterYolo+=1
    elif (dataYolo[counterYolo,0]>dataOF[counterOF,0]):
        # if data is -1 -> repace previous data
        for i  in range(dataOF.shape[1]):
            if dataOF[counterOF,i] == -1:
                if counterIteration != 0:
                    dataOF[counterOF,i] = data[counterIteration-1][i]
        data.append(dataOF[counterOF,:].tolist())
        counterOF+=1
    elif (dataYolo[counterYolo,0]<dataOF[counterOF,0]):
        # if data is -1 -> repace previous data
        for i  in range(dataOF.shape[1]):
            if dataOF[counterOF,i] == -1:
                if counterIteration != 0:
                    dataOF[counterOF,i] = data[counterIteration-1][i]
        data.append(dataOF[counterOF,:].tolist())
        counterYolo+=1
        counterOF+=1
    counterIteration += 1
data = np.array(data)
print(data)
```

# Evaluate data

```python
counterGT = 0
counterRT = 0
counterMatch = 0
errorLS = 0
errorsLS = []
errorRS = 0
errorsRS = []
errorLE = 0
errorsLE = []
errorRE = 0
errorsRE = []
errorLW = 0
errorsLW = []
errorRW = 0
errorsRW = []
def calculateError(data1,data2):
    return ((data1[1]-data2[1])**2+(data1[2]-data2[2])**2)**(1/2)
while (counterGT<dataGT.shape[0] and counterRT<data.shape[0]):
    if (dataGT[counterGT,0]==data[counterRT,0]):
        frameIndex = dataGT[counterGT,0]
        #LS
        errLS = calculateError(dataGT[counterGT,:3],data[counterRT,:3])
        errorsLS.append([frameIndex,errLS])
        errorLS += errLS
        #RS
        errRS = calculateError(dataGT[counterGT,3:6],data[counterRT,3:6])
        errorsRS.append([frameIndex,errRS])
        errorRS += errRS
        #LE
        errLE = calculateError(dataGT[counterGT,6:9],data[counterRT,6:9])
        errorsLE.append([frameIndex,errLE])
        errorLE += errLE
        #RE
        errRE = calculateError(dataGT[counterGT,9:12],data[counterRT,9:12])
        errorsRE.append([frameIndex,errRE])
        errorRE += errRE
        #LW
        errLW = calculateError(dataGT[counterGT,12:15],data[counterRT,12:15])
        errorsLW.append([frameIndex,errLW])
        errorLW += errLW
        #RW
        errRW = calculateError(dataGT[counterGT,15:18],data[counterRT,15:18])
        errorsRW.append([frameIndex,errRW])
        errorRW += errRW
        # increment counter
        counterMatch += 1
        counterRT += 1
        counterGT += 1
    elif (dataGT[counterGT,0]>data[counterRT,0]):
        counterRT +=1
    elif (dataGT[counterGT,0]<data[counterRT,0]):
        counterGT +=1
errorLS /= counterMatch
errorRS /= counterMatch
errorLE /= counterMatch
errorRE /= counterMatch
errorLW /= counterMatch
errorRW /= counterMatch
print(f"Error (RMSE) ::\nLS : {round(errorLS,1)} pixel, RS : {round(errorRS,1)} pixel\nLE : {round(errorLE,1)} pixel, RE : {round(errorRE,1)} pixel\nLW : {round(errorLW,1)} pixel, RW : {round(errorRW,1)} pixel\n")
```
